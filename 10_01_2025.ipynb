```python
def dev_dec(func):
    def inner(a, b):
        if b == 0:
            return "Cannot division by zero."
        if a<b:
            a,b=b,a
        return func(a, b)
    return inner
@dev_dec
def div(a, b):
    return a / b
print(div(2, 4))
print(div(4, 2))
print(div(4, 0))
```

```python
def dev_dec(func):
    def inner(a, b):
        if a * b > 20:
            return a / b
        else:
            return func(a, b)
    return inner
@dev_dec
def mul(a, b):
    return a * b
print(mul(2, 4))
print(mul(4, 2))
print(mul(4, 20))
```
